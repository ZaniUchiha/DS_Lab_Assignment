```python
class Stack:
    def __init__(self, size=100):
        self.data = []         # Make a box to keep things
        self.size = size       # How many things can fit in the box

    def push(self, val):
        if len(self.data) >= self.size:   # Is the box full?
            raise IndexError("Stack overflow") # No more space!
        self.data.append(val)             # Put a thing in the box

    def pop(self):
        if len(self.data) <= 0:           # Is the box empty?
            raise IndexError("Stack Underflow") # Nothing to take out!
        return self.data.pop()            # Take out the top thing

    def is_empty(self):
        return len(self.data) == 0        # Is the box empty?

    def peek(self):
        if self.is_empty():               # Is the box empty?
            raise IndexError("Stack is empty") # Nothing on top!
        return self.data[-1]              # Show the top thing


class InfixToPostfix:
    def __init__(self):
        self.prec = {
            "^": 3,      # Power is strongest
            "*": 2,      # Times is next
            "/": 2,      # Divide is next
            "+": 1,      # Plus is weak
            "-": 1       # Minus is weak
        }
        self.right_assoc = "^"  # Power goes from right to left
        self.stack = Stack()    # Make a box for signs

    def is_operand(self, ch):
        return ch.isalnum()     # Is it a letter or number?

    def convert(self, infix: str) -> str:
        out = []                # This will hold the answer
        self.stack = Stack()    # Make a new box for each math sentence

        for ch in infix.replace(" ", ""):  # Look at each thing in the math sentence
            if self.is_operand(ch):        # If it's a letter or number
                out.append(ch)             # Put it in the answer

            elif ch == "(":                # If it's a (
                self.stack.push(ch)        # Put it in the box

            elif ch == ")":                # If it's a )
                while not self.stack.is_empty() and self.stack.peek() != "(":
                    out.append(self.stack.pop()) # Take out signs until (
                if self.stack.is_empty():
                    raise ValueError("Missing opening parenthesis") # No matching (
                self.stack.pop()           # Throw away the (

            else:  # If it's a math sign
                while (not self.stack.is_empty()
                       and self.stack.peek() != "("
                       and ((self.prec[self.stack.peek()] > self.prec[ch])
                            or (self.prec[self.stack.peek()] == self.prec[ch]
                                and ch != self.right_assoc))):
                    out.append(self.stack.pop()) # Take out stronger signs
                self.stack.push(ch)              # Put the new sign in the box

        # Take out any signs left in the box
        while not self.stack.is_empty():
            top = self.stack.pop()
            if top == "(":
                raise ValueError("Missing closing parenthesis") # No matching )
            out.append(top)

        return "".join(out)           # Make the answer a string

conv = InfixToPostfix()               # Make a helper

expr1 = "A+B*C"                       # Math sentence
expr2 = "(A+B)*C"
expr3 = "A+B*C-D/E"
expr4 = "A*(B+C)/D"
expr5 = "A^B^C"

for expr in [expr1, expr2, expr3, expr4, expr5]:
    result = conv.convert(expr)        # Change to postfix
    print(f"{expr} -> {result}")       # Show the answer
```

```python
class Stack:
    def __init__(self, size):
        self.data = []              # Make a box to keep things
        self.max_size = size        # How many things can fit in the box

    def push(self, val):
        if len(self.data) >= self.max_size:   # Is the box full?
            raise IndexError("Stack overflow") # If yes, say "No more space!"
        self.data.append(val)                 # Put a thing in the box

    def pop(self):
        if len(self.data) <= 0:               # Is the box empty?
            raise IndexError("Stack Underflow")# If yes, say "Nothing to take out!"
        return self.data.pop()                # Take out the top thing

    def is_empty(self):
        return len(self.data) == 0            # Is the box empty? Yes or No

    def peek(self):
        if self.is_empty():                   # Is the box empty?
            raise IndexError("Stack is empty")# If yes, say "Nothing on top!"
        return self.data[-1]                  # Show the top thing

    def size(self):                           # How many things are in the box now
        return len(self.data)


class PostFixEvaluator:
    def __init__(self):
        self.stack = Stack(100)               # Make a box that can hold 100 things
    
    def apply(self, op, a, b):
        if op == "+": return a + b            # Add numbers
        if op == "-": return a - b            # Subtract numbers
        if op == "*": return a * b            # Multiply numbers
        if op == "/": return a / b            # Divide numbers
        if op == "^": return a ** b           # Power (a to the b)
        raise ValueError(f"Unknown operator {op}") # If not a known sign, say "bad sign"
    
    def evaluate(self, postfix: str) -> float:
        for ch in postfix.replace(" ", ""):   # Look at each thing in the math sentence
            if ch.isdigit():                  # If it's a number
                self.stack.push(float(ch))    # Put the number in the box
            elif ch in "+-*/^":               # If it's a math sign
                if self.stack.size() < 2:     # Need two numbers to do math
                    raise ValueError("Malformed expression: insufficient operands")
                b = self.stack.pop()          # Take out one number
                a = self.stack.pop()          # Take out another number
                self.stack.push(self.apply(ch, a, b)) # Do the math and put answer back
            else:
                raise ValueError(f"Bad token {ch}")   # If it's not a number or sign, say "bad thing"
        
        if self.stack.size() != 1:            # Should be only one answer left
            raise ValueError("Malformed expression: leftover values")
        return self.stack.pop()               # Give back the answer


ev = PostFixEvaluator()                       # Make a math helper
expr1 = "432+*"                              # Math sentence: 4*(3+2)
res1 = ev.evaluate(expr1)                    # Get the answer
print(f"{expr1} -> {res1}")                  # Show the answer
```

class linearSerch:
    def __init__(self, data):
        self.data = list(data)  # Save the numbers in a list

    def find(self, target):
        for i, x in enumerate(self.data):  # Look at each number one by one
            if x == target:                # If the number is what we want
                return i                   # Tell where it is
        return -1                          # If not found, say "not here"
